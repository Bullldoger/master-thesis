```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 10
n = 100
x_min, x_max = -L, L
dx = (x_max - x_min) / n
x = np.arange(x_min, x_max + dx / 2, dx)
n = x.shape[0]
```

```python
lower_diag = np.diag(-1 / (dx ** 2) * np.ones(n - 1), k = -1)
upper_diag = np.diag(-1 / (dx ** 2) * np.ones(n - 1), k = 1)
diag       = np.diag(2 / (dx ** 2) * np.ones(n) + x ** 2, k = 0)

D = lower_diag + upper_diag + diag

D[0]  = 0
D[-1] = 0

D[0, 0]   = 0
D[-1, -1] = 0
```

```python
eig_pairs = []
eigs_vls, eigs_vc = np.linalg.eig(D)

for eig_val, eig_vec in zip(eigs_vls, eigs_vc):
    eig_pairs += [(eig_val, eig_vec)]
    
srt_eigs_pairs = sorted(eig_pairs, key=lambda x: x[0])[2:]
```

## plot show the first ten numerically found eigenvalues Î» together with their theoretical value

```python
plt.figure(figsize=(13, 8))
plt.plot(list(range(1, 11, 1)), [e for e, v in srt_eigs_pairs[:10]], 'or', label='Numerical eigenvalues', markersize=4)
plt.plot(list(range(1, 11, 1)), list(range(1, 21, 2)), 'bo', label='Numerical eigenvalues', markersize=4)

plt.xlim([0, 10])
plt.ylim([0, 20])
plt.xticks(list(range(0, 12, 1)))
plt.yticks(list(range(0, 22, 1)))
plt.legend()
plt.grid()
plt.show()
```

##  Plot the eigenvectors that correspond to the lowest five eigenvalue

```python
plt.figure(figsize=(13, 8))
for i in range(5):
    plt.plot(srt_eigs_pairs[i][1].flatten(), label='Eigen value: {}'.format(srt_eigs_pairs[i][0]), markersize=3)
    
plt.grid()
plt.legend()
plt.show()
```
