```python
import torch
import torch.nn as nn
import torch.optim as optim

import numpy as np

import matplotlib.pyplot as plt
```

```python
import sympy as sm
from sympy.utilities.lambdify import lambdify, implemented_function
from sympy import fourier_series, pi
from sympy.abc import x
```

```python
dim = 100
A = np.random.randn(dim, dim)
x = np.random.randn(dim, 1)
b = A @ x
b_e = np.random.randn(dim, 1) / 25

norm  = lambda x: np.linalg.norm(x)
kappa = lambda x: norm(x) * norm(np.linalg.inv(x))

x_ = np.linalg.inv(A) @ (b + b_e)
```

```python
plt.figure(figsize=(16, 5))
plt.subplot(1, 2, 1)
plt.title('b values', fontsize=18)
plt.scatter(b, b + b_e, 5, label='Values with small error')
plt.plot(b, b, '--r', linewidth=0.5, alpha=0.8, label='Values without error')
plt.xlabel('b', fontsize=14)
plt.ylabel('b', fontsize=14)
plt.grid()
plt.legend(fontsize=14)

plt.subplot(1, 2, 2)
plt.title('x values', fontsize=18)
plt.scatter(x, x_, 5, label='Values with error')
plt.plot(x, x, '--r', linewidth=0.5, alpha=0.8, label='Values without error')
plt.xlabel('x', fontsize=14)
plt.ylabel('x', fontsize=14)
plt.grid()
plt.legend(fontsize=14)

plt.show()
```
