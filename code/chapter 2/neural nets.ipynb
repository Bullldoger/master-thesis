```python
import daft
from matplotlib import rc

rc("font", family="serif", size=12)
rc("text", usetex=True)
```

```python
p_color = {"ec": "#46a546"}
s_color = {"ec": "#f89406"}
```

```python
pgm = daft.PGM(dpi=200)
```

```python
x = pgm.add_node('x', 'x', 0, 1.0, fontsize=15, plot_params=s_color)
y = pgm.add_node('y', 'y', 3.0, 1.0, fontsize=15, plot_params=s_color)
z = pgm.add_node('z', 'z', 6.0, 1.0, fontsize=15, plot_params=p_color)
```

```python
pgm.add_plate([2.0, 0.25, 2.0, 1.5], r'$y = \sigma\left(A x + b \right) $', position='bottom center', label_offset=[0, 3], fontsize=14)
```

```python
exy = pgm.add_edge('x', 'y', label='Affine transform', xoffset=-0.5, plot_params=s_color, label_params={'fontsize': 10}) 
eyz = pgm.add_edge('y', 'z', label='Affine transform', xoffset=0.5, plot_params=s_color, label_params={'fontsize': 10}) 
```

```python
pgm.render()
```
