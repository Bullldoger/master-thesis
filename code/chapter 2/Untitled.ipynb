```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from utilities.utilities import *
```

```python
size = 10
randomly_points = np.random.rand(size ** 2, 2)
points = get_points(size=size).cpu().detach().numpy()
```

```python
plt.figure(figsize=(10, 3))

plt.subplot(1, 2, 1)
plt.title('Fixed points', fontsize=18)
plt.scatter(points[:, 0], points[:, 1])
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

plt.subplot(1, 2, 2)
plt.title('Randomly choosen points', fontsize=18)
plt.scatter(randomly_points[:, 0], randomly_points[:, 1])
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

plt.show()
```
